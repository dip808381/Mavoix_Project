```python
df2 = df.loc[df["Topic"] == "Eye Diseases and Injury"]
df2.head()
```

```python
Eye_Diseases_Injury1 = df2["LocationDesc"].value_counts()
Eye_Diseases_Injury = Eye_Diseases_Injury1.to_frame()
Eye_Diseases_Injury
```

```python
### type your answer here
# generate plot
Eye_Diseases_Injury1.sort_values().plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Person')
plt.title('Eye Diseases and Injury')
plt.show()
```

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

stopwords = set(STOPWORDS)

alice_novel =  " ".join(review for review in Eye_Diseases_Injury1.index)
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
df3 = df.loc[df["Question"] == "Percentage of Adults 18 Years and Older Who Reported Blindness or Severe Difficulty Seeing"]
df3.head()
```

```python
Adults_18_Years = df3["LocationDesc"].value_counts()
Adults_18_Years.to_frame().head()

Adults_18_Years.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Place name') # add to x-label to the plot
plt.ylabel('Number of patients') # add y-label to the plot
plt.title('Percentage of Adults 18 Years and Older Who Reported Blindness or Severe Difficulty Seeing') # add title to the plot

plt.show()
```
